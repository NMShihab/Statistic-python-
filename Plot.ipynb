# Plot Histrogram

```python
# Set default Seaborn style
sns.set()
```

**Simple plot of histogram**

```python
# Function that will plot a histogram 
def histogram(df,column,xlabel):
    # Plot histogram
    plt.hist(df[column])
        
    # Label axes
    plt.xlabel(xlabel)
    plt.ylabel('count')
      
    # Show histogram
    plt.show()  
    
```

```python
# Plot histogram of versicolor petal lengths
histogram(versicolor,'Petal length (cm)','petal length (cm)')
```

```python
# Plot histogram of Setosa petal lengths
histogram(setosa,'Petal length (cm)','petal length (cm)')
```

```python
# Plot histogram of Virginica petal lengths
histogram(virginica,'Petal length (cm)','petal length (cm)')
```

**Add bins**

```python
# Function that will plot a histogram with define bins
def histogram_with_bins(df,column,xlabel):
    # Compute number of data points: n_data
    n_data = len(df[column].values)
    
    # Number of bins is the square root of number of data points: n_bins
    n_bins = np.sqrt(n_data)

    # Convert number of bins to integer: n_bins
    n_bins = int(n_bins)
    
    # Plot the histogram
    plt.hist(df[column],bins=n_bins)
    
    # Label axes
    plt.xlabel(xlabel)
    plt.ylabel('count')


    # Show histogram
    plt.show()
    
    
    
```

```python
# Plot histogram of versicolor petal lengths
histogram_with_bins(versicolor,'Petal length (cm)','petal length (cm)')
```

```python
# Plot histogram of Setosa petal lengths
histogram_with_bins(setosa,'Petal length (cm)','petal length (cm)')
```

```python
# Plot histogram of Virginica petal lengths
histogram_with_bins(virginica,'Petal length (cm)','petal length (cm)')
```

# Bee swarm plot

```python
# Create bee swarm plot with Seaborn's default settings
_ = sns.swarmplot(x='variety', y='Petal length (cm)', data=iris)

# Label the axes
plt.xlabel('species')
plt.ylabel('petal length (cm)')

# Show the plot
plt.show()
```

```python
def Swarm_plot(data,x,y):
    _ = sns.swarmplot(x=x, y=y, data=data)

    # Label the axes
    plt.xlabel(x)
    plt.ylabel(y)

    # Show the plot
    plt.show()
    
```

```python
# Plot swarm_plot of petal length and variety
Swarm_plot(data=iris,x='variety',y='Petal length (cm)')
```

```python
# Plot swarm_plot of petal width and variety
Swarm_plot(data=iris,x='variety',y='Petal width (cm)')
```

```python
# Plot swarm_plot of sepal length and variety
Swarm_plot(data=iris,x='variety',y='Sepal length (cm)')
```

```python
# Plot swarm_plot of sepal width and variety
Swarm_plot(data=iris,x='variety',y='Sepal width (cm)')
```
